```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from torch.nn.functional import softmax
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
model_name = "RashidNLP/Amazon-Deberta-Base-Sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)
    probs = softmax(outputs.logits, dim=1)
    labels = ["Negative", "Neutral", "Positive"]
    return dict(zip(labels, probs[0].tolist()))
```

```python
text=input()
labels = ["sarcasm", "not sarcastic"]
result = classifier(text, candidate_labels=labels)
result2=predict_sentiment(text)
#print(result['scores'])
if result2['Negative']>result2['Positive'] and result2['Negative']>result2['Neutral']:
  res="Negative"
elif result2['Positive']>result2['Negative'] and result2['Positive']>result2['Neutral']:
  res="Positive"
else:
  res="Neutral"
if result['scores'][0]-result['scores'][1]>0.15:
  if res=='Positive':
    print("Negative")
  elif res=='Negative':
    print("Positive")
  else:
    print("Neutral")
elif result['scores'][1]-result['scores'][0]<0.15:
    print("Neutral")
else:
  print(res)
```
